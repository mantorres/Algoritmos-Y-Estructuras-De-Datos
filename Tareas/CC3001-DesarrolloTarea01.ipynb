#### Objetivo del trabajo
El objetivo de esta tarea es implementar una función ``palabras(n)`` que al ser llamada entregue un string que contenga el número ``n`` expresado en palabras, donde ``n`` es un número entero nonegativo menor que mil millones.

Por ejemplo,

```
palabras(12) = "doce"
palabras(41025) = "cuarenta y un mil veinticinco"
```

Como primer paso, comenzaremos escribiendo una función restringida a números menores que mil:

````python
"""=============================================================================
palabras1000:
el valor que ingresa es un valor entero n entre 0 y 999, la funcion retorna un
string con el nombre del numero n
============================================================================="""
def palabras1000(n):
  # filtro
  assert type(n)==int
  assert 0<=n<=999
  
  # definicion de los valores en string para los casos generales
  unidades = ["cero","un","dos","tres","cuatro","cinco",
              "seis","siete","ocho","nueve"]
  decenas = ["","diez","veinte","treinta","cuarenta","cincuenta",
             "sesenta","setenta","ochenta","noventa"]
  centenas = ["","ciento","doscientos","trescientos","cuatrocientos",
              "quinientos","seiscientos","setecientos",
              "ochocientos","novecientos"]
  # escribir n en base 10 por coordenada en un vector
  (c,d,u)=(n//100,(n%100//10),n%10)

  # correccion del 1 y 100
  if c==1 and d==0 and u==0:
    return "cien"
  elif c==0 and d==0 and u==1:
    return "uno"

  # correcciones cuando decena es cero:
  # este caso no considera el 1 a estas alturas del algoritmo
  elif d==0:
    if c==0:
      return unidades[u]
    if u==0:
      return centenas[c]
    if c!=0:
      return centenas[c]+" "+unidades[u]

  # correcciones cuando la unidad es cero
  elif u==0:
    return centenas[c]+" "+decenas[d]

  # correccion a los numeros entre 10 y 15
  # debido a que se utilizan palabras especificas
  elif d==1 and (u==0 or u==1 or u==2 or u==3 or u==4 or u==5):
    if u==0:
      return centenas[c]+" diez"
    if u==1:
      return centenas[c]+" once"
    if u==2:
      return centenas[c]+" doce"
    if u==3:
      return centenas[c]+" trece"
    if u==4:
      return centenas[c]+" catorce"
    if u==5:
      return centenas[c]+" quince"
  
  # cuando no se dan los casos particulares anteriores
  else:
    return centenas[c]+" "+decenas[d]+" y "+unidades[u]
````

A continuación vemos el caso general:

````python
"""=============================================================================
palabras:
-el valor que ingresa es un valor entero n entre 0 y 999999999, la funcion 
retorna un string con el nombre del numero n.
-hay 3 casos luego de realizar una descomposicion (a,b,c) descrita 
en el algoritmo, segun la cantidad de digitos significativos del entero n, la 
distincion de estos 3 casos permiten llamar menos veces a la funcion 
palabras1000 cuando no sea necesaria
-en los casos donde n>=1000, se distingue cuando solo hay una unidad de mil o 
una unidad de millon.
============================================================================="""
def palabras(n):
  assert 0<=n<1000000000

  # caso base
  if n==0:
    return "cero"
  """
  DESCOMPOSICION (a,b,c)
  el siguente vector tiene en cada coordenada un numero entre 0 y 999
  tal que sus digitos corresponden respectivamente a los digitos de n como:
  a = centena- decena- unidad de millones
  b = centena- decena- unidad de miles
  c = centena- decena- unidad (c cita a palabras1000uno)
  """
  (a,b,c)=(n//1000000,(n%1000000)//1000,n%1000)

  # segun el valor de n podemos llamar menos veces a la funcion palabras1000
  # separando principalmente en 3 casos debido a la descomposicion (a,b,c)

  """
  0<=n<=999
  """
  # el primer caso es aplicar palabras1000 de forma simple
  if n<=999:
    return palabras1000uno(c)

  # el segundo y tercer caso tienen subcasos donde se particulariza cuando hay
  # ceros a la derecha de digitos no nulos, y ademas se hace distincion de los
  # casos singulares, estos son los que hay mil o un millon

  """
  1000<=n<=999999
  """
  if 1000<=n<=999999:
    if b==1:
      # cuando b=1, se tiene mil y algo, que se denota en singular
      if c==0:
        return "mil"
      else:
        return "mil "+palabras1000uno(c)
    elif c==0:
      return palabras1000(b)+" mil"
    else:
      return palabras1000(b)+" mil "+palabras1000uno(c)

  """
  1000000<=n<=999999999
  """
  if 1000000<=n:
    # cuando a=1, se tiene un millon, que se denota en singular
    if a==1:
      if b==0 and c==0:
        return "un millon"
      # sub caso con el 1000
      elif b==1:
        if c==0:
          return "un millon mil"
        else:
          return "un millon mil "+palabras1000uno(c)
      elif b==0:
        return "un millon "+palabras1000uno(c)
      elif c==0:
        return "un millon "+palabras(b)+" mil"
      else:
        return "un millon "+palabras1000(b)+" mil "+palabras1000uno(c)
    # cuando hay ceros a la derecha de los digitos de millones
    # se consideran varios casos
    elif b==0:
      if c==0:
        return palabras1000(a)+" millones "
      else:
        return palabras1000(a)+" millones "+palabras1000uno(c)
    elif c==0:
      return palabras1000(a)+" millones "+palabras1000(b)+" mil"
    # cualquier otro caso
    else:
      return palabras1000(a)+" millones "+palabras1000(b)+" mil "+palabras1000uno(c)
````

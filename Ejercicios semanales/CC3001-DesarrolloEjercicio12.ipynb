### Ejercicio 7.3 (Merge de listas con repeticiones)

Suponga que se desea mezclar dos listas ordenadas de manera ascendente en las cuales puede haber elementos repetidos. Estas listas se representan en forma compacta como una secuencia de tuplas en que para cada elemento se indica su *multiplicidad*, es decir, el número de veces que se repite. Por ejemplo, la lista
```
[(12,2),(34,1),(56,3),(74,1),(81,1)]
```
es la representación compacta de la lista
```
[12,12,34,56,56,56,74,81]
```
Se pide escribir un método ``merge(a,b)`` que reciba como argumento dos listas compactas ``a`` y ``b`` y retorne el resultado de mezclarlas. El resultado debe estar también en formato compacto.

Para esto, usted debe modificar el método ``merge``del apunte, el que aparece a continuación:


````python
def mergesimple(a,b):
    """-------------------------------------------------------------------------
    Funcion Merge original.
    -------------------------------------------------------------------------"""
    i=0
    j=0
    while i<len(a) or j<len(b):
        if j>=len(b) or (i<len(a) and a[i]<=b[j]):
            yield a[i]
            i=i+1
        else:
            yield b[j]
            j=j+1

def merge(a,b):
    """-------------------------------------------------------------------------
    ### Descompactificar las listas a,b:
    Descripcion: Se crea una lista vacia. Se recorre la lista a descompactificar 
    y se añade el elemento i-esimo tantas veces como su contador. La informacion
    se encuentra en formato tupla como: (elemento, contador).
    -------------------------------------------------------------------------"""
    # Listas vacias para luego añadir
    A = []; B = []  
    # Descompactificar las listas a, b en las listas A,B respectivamente
    for i in range(len(a)):                 # Descompactificar primera lista
        for j in range(a[i][1]):            
            A.append(a[i][0])
    for i in range(len(b)):                 # Descompactificar segunda lista
        for j in range(b[i][1]):       
            B.append(b[i][0])
    """-------------------------------------------------------------------------
    ### Meclar ambas listas no compactas:
    Comentario: Para tener el codigo mas ordenado como en esta parte se utiliza
    la mezcla del merge original, el cual deje como una funcion auxiliar
    -------------------------------------------------------------------------"""
    AB = [x for x in mergesimple(A,B)]  # Mezclar A con B
    """-------------------------------------------------------------------------
    ### Compactificar mezcla AB que contiene la mezcla de A y B:
    Descripcion: Se crea una lista vacia C. Se recorre la lista AB viendo su 
    elemento i-esimo, si este elemento es igual al siguiente se aumenta el
    contador, en caso que el elemento i-esimo no sea igual al siguiente se crea
    la tupla y se guarda la informacion del elemento con su contador. Se añade
    un caso de parada para el elemento len(AB)-1-esimo.
    -------------------------------------------------------------------------"""
    C = []                      # Arreglo para almacenar forma compacta
    count = 1                   # Contador de repeticiones del elemento actual
    for i in range(len(AB)):    # Recorrer el arreglo AB
        if i == len(AB) - 1:    # Caso final
            c = (AB[i], count)  # Crear la tupla y guardar la info en ella
            C.append(c)         # Añadir la tupla a la lista compacta
        elif AB[i] == AB[i+1]:  # Cuando hay repeticion se aumenta el contador
            count = count + 1   # Aumentar el contador
        elif AB[i] != AB[i+1]:  # Caso en que no se repite el elemento
            c = (AB[i], count)  # Crear la tupla y guardar la info en ella
            C.append(c)         # Añadir la tupla a la lista compacta
            count = 1           # Reiniciar el contador
    return C
````

````python
a=[(12,2),(34,1),(56,3),(74,1),(81,1)]
b=[(10,3),(12,5),(65,1),(74,1),(90,3)]
print(merge(a,b))
````
